```python
# 필요한 라이브러리를 가져오기
import matplotlib.pyplot as plt  # Matplotlib를 가져옴
import pandas as pd  # Pandas를 가져옴
import FinanceDataReader as fdr  # FinanceDataReader를 가져옴
```

```python
# 첫 번째 인자로 종목 코드 '005930'을 전달하고, 두 번째 인자로 시작일을 '2015-01-01'로 지정합니다.
df = fdr.DataReader('005930', '2025-01-01')

# DataFrame의 head() 메서드를 사용하여 처음 5개 행을 표시합니다.
df.head(5)
```

```python
# 주식 데이터프레임(df)에서 'Close' 열의 데이터를 추출하고, 이를 사용하여 일별 수익률을 계산하는 코드입니다.

# .pct_change(): 일별 수익률을 계산하는 과정으로, 현재 값과 이전 값 사이의 백분율 변화를 나타냅니다( 계산된 백분율 변화율을 백분율로 표시하기 위해 100을 곱합니다)

df = (df['Close'].pct_change()) * 100
```

```python
df.describe()
```

```python
# plt.figure(figsize=(12, 6)): 그래프의 크기를 설정합니다. 가로 12, 세로 6 인치 크기로 설정합니다.
plt.figure(figsize=(12, 6))

# plt.plot(r.index, r, label='Daily Returns', color='b', linestyle='-'):
# 'df' 데이터를 이용하여 그래프를 그립니다.
# df.index: 'r' 데이터의 인덱스(날짜)를 x 축으로 사용합니다.
# df: 일별 수익률 데이터를 y 축으로 사용합니다.
# color='b': 그래프의 색상을 파란색(blue)으로 설정합니다.
# linestyle='-': 그래프의 선 스타일을 실선으로 설정합니다.
plt.plot(df.index, df,color='b', linestyle='-')

# 그래프의 제목을 설정합니다.
plt.title('Daily Returns of Samsung Electronics (005930) Stock')

#  x 축의 레이블을 설정합니다.
plt.xlabel('Date')

# y 축의 레이블을 설정합니다.
plt.ylabel('Daily Returns (%)')

# 그래프에 그리드를 활성화합니다.
plt.grid(True)

# 그래프를 표시합니다.
plt.show()
```

```python
import matplotlib.pyplot as plt

# 히스토그램 그래프를 그리기 위한 일별 수익률 데이터 'df'를 사용합니다.
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.hist(df, bins=20, color='skyblue', edgecolor='black')  # 히스토그램 그리기
plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Returns (%)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
